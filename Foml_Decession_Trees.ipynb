```python
def predict(root,X_test):
  def predict1(node, data_record):
      # Traverse the tree until a leaf node is reached
      while node.left is not None or node.right is not None:
          # Get the feature value from the data record


          if node.feature is None or node.threshold is None:
              return node.value
              # raise ValueError(f"Feature '{node.feature}' not found in data record.")
          feature_value = data_record[node.feature]
          # Compare the feature value with the node's threshold
          if feature_value <= node.threshold:
              node = node.left  # Move to the left subtree
          else:
              node = node.right  # Move to the right subtree

      # Return the value at the leaf node
      return node.value
  classification_results = []
  for index, record in X_test.iterrows():
      result = predicted_value = predict1(root, record)
      classification_results.append(result)
  return classification_results
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score  # You can choose other metrics as well
import numpy as np


def kfold_cross_validation( X, y, k=10):
    # Initialize KFold with the number of splits
    kf = KFold(n_splits=k, shuffle=True, random_state=42)

    # List to store the performance metric for each fold
    results = []

    # Perform k-fold cross-validation
    for train_index, test_index in kf.split(X):
        # Split data into training and test sets
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        root=Node(X_train)
        # Train the model on the training set
        # model.fit(X_train, y_train)
        root= DecessionTreeClassifier(0,root)
        # Predict on the test set
        y_pred = predict(root,X_test)

        # Calculate performance metric (accuracy in this case)
        accuracy = accuracy_score(y_test, y_pred)

        # Store the result
        results.append(accuracy)

    # Calculate the average performance across all folds
    average_performance = np.mean(results)

    return average_performance, results
```

```python
def calculate_height(node):
    if node is None:
        return 0
    else:
        # Recursively calculate the height of left and right subtrees
        left_height = calculate_height(node.left)
        right_height = calculate_height(node.right)

        # The height of the tree is the maximum height of its subtrees plus 1 for the current node
        return max(left_height, right_height) + 1
```
